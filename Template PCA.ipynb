## Import Libraries 

```python
import pandas as pd
import numpy as np

np.set_printoptions(suppress=True)
```

## Import the dataset

```python
df= pd.read_csv('GDL_Indicators_(2019)_data.csv')
df
```

## Define the dataset

```python
X = df[['Life expectancy','Mean years schooling','Population size in thousands','GNI per capita in thousands of US$ (2011 PPP)']].values
X
```

## Scale the data between o-1

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_scaled = sc_X.fit_transform(X)
X_scaled
```

## Use PCA to reduce dimensionality

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
X_scaled = pca.fit_transform(X_scaled)
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
print(pca.components_.T * np.sqrt(pca.explained_variance_))
```
